```python
vaccine_data = []
for vaccine_chart_type in [co for co in y['data'] if co not in ['Updated On', 'State']]:
    vaccine_data.append({
        "labels": y['data']['Updated On'],
        "datasets": [{
            "data": y['data'][vaccine_chart_type],
            "label": vaccine_chart_type
        }]
    })
```

```python
vaccine_data
```

```python
ff = []
ff.append({
    "labels": y['data']['Updated On'],
    "datasets": [j[i][0] for j in vaccine_data for i in ['Male', 'Female', 'Transgender'] if i in j]
})

ff.append({
    
    "labels": y['data']['Updated On'],
    "datasets": [j[i][0] for j in vaccine_data for i in ['delta_male', 'delta_female', 'delta_transgender'] if i in j]
})
```
