```python
df= pd.DataFrame(data)
```

```python
df = df.sort_values(by='created_at', ascending=True)
df.drop(['_id', 'agg_by'], axis=1, inplace=True)
df.head()
```

```python
df['Category'] = df['category'].apply(lambda x: x[0] if x and len(x)>=1 else "NA")
df['SubCategory'] = df['category'].apply(lambda x: x[1] if x and len(x)>1 else "NA")
```

```python
df.tail()
```

```python
df['created_at'] = df['created_at'].dt.date
```

```python
df['Category'] = df['Category'].apply(lambda x: x.replace('#available', 'Available'))
```

```python
samp = df.groupby(['created_at', 'Category']).agg({'text': 'count'}).reset_index()
samp2 = df.groupby(['created_at', 'SubCategory']).agg({'text': 'count'}).reset_index()
```
