```python
import requests
import json
import pandas as pd
import time
from tqdm import tqdm

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', -1)
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
base_link = 'https://cdn-api.co-vin.in/api'
```

```python
API_DICT = {
    "state_list": "/v2/admin/location/states",
    "district_list": "/v2/admin/location/districts/", # /{state_id},
    "appointment_by_pin": "/v2/appointment/sessions/public/findByPin", #?pincode=xxxxx&date=xx-xx-xxxx
    "appointment_by_district": "/v2/appointment/sessions/public/findByDistrict", #?district_id=xxx&date=xx-xx-xxxx
    "7_day_appointment_by_pin": "/v2/appointment/sessions/public/calendarByPin", #?pincode=xxxxx&date=xx-xx-xxxx
    "7_day_appointment_by_district": "/v2/appointment/sessions/public/calendarByDisctrict", #?district_id=xxx&date=xx-xx-xxxx   
}
```

```python
state_link = base_link + API_DICT['state_list']
resp = requests.get(state_link, headers=header)
```

```python
state_list = json.loads(resp.text)
state_list = state_list['states']
```

```python
print(state_list)
```

```python
dist_dict = {}
for st in tqdm(state_list):
    dist_link = base_link + API_DICT['district_list'] + str(st['state_id'])
    dist_resp = requests.get(dist_link, headers=header)
    dist_dict[st['state_name']] = json.loads(dist_resp.text)
    time.sleep(4)
```
