```python
dff1 = pd.read_csv(f"{BASE_PATH}\\twitter_db\\kaggle_data\\time_series_covid_19_confirmed.csv")
dff2 = pd.read_csv(f"{BASE_PATH}\\twitter_db\\kaggle_data\\time_series_covid_19_deaths.csv")
dff3 = pd.read_csv(f"{BASE_PATH}\\twitter_db\\kaggle_data\\time_series_covid_19_recovered.csv")
```

```python
dff1[dff1['Country/Region'] == "United Kingdom"]
```

```python
dff1.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
dff2.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
dff3.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
```

```python
dff1 = dff1.groupby(['Country/Region']).sum()
dff2 = dff2.groupby(['Country/Region']).sum()
dff3 = dff3.groupby(['Country/Region']).sum()

dff1.reset_index(inplace=True)
dff2.reset_index(inplace=True)
dff3.reset_index(inplace=True)
```

```python
samp1 = dff1[dff1["Country/Region"]=="USA"].T
samp2 = dff2[dff2["Country/Region"]=="China"].T
samp3 = dff3[dff3["Country/Region"]=="China"].T
```

```python
samp1.head()
```

```python
samp1.reset_index(inplace=True)
samp2.reset_index(inplace=True)
samp3.reset_index(inplace=True)

samp1.columns= ['Date', 'Confirmed']
samp2.columns= ['Date', 'Deaths']
samp3.columns= ['Date', 'Recovered']
```

```python
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['Date']), [samp1, samp2,samp3])
```

```python
df_merged = df_merged[4:]
```

```python
df_merged.dropna(axis=0, inplace=True)
```

```python
df_merged['Confirmed'] = df_merged['Confirmed'].astype('int')
df_merged['Deaths'] = df_merged['Deaths'].astype('int')
df_merged['Recovered'] = df_merged['Recovered'].astype('int')
```

```python
df_merged.head()
```

```python
diff_df = df_merged.set_index('Date').diff().reset_index()
```
